```python
from transformers import AutoTokenizer, AutoModelForCausalLM

# Replace 'path/to/your/model' with the actual path to your local directory
model_dir = r"models/model"

# Load the tokenizer (assuming it's named 'tokenizer.json' in the directory)
tokenizer = AutoTokenizer.from_pretrained(model_dir)
```

```python
# Load the model (assuming it's named 'pytorch_model.bin' in the directory)
model = AutoModelForCausalLM.from_pretrained(model_dir)
```

```python
model.load_adapter('models/adapter')
```

```python
# Define the product_info string with proper escaping
product_info = """{'Product_Company_Name': 'Apple', 'Product_Category': 'Electronics , Laptops', 'Product_Name': 'MacBook Air M1', 'Product_Description': 'The MacBook Air M1 is a powerful and portable laptop that is perfect for students, professionals, and anyone else who wants a powerful computer that they can take with them wherever they go. It features a stunning Retina display, a long-lasting battery, and a powerful M1 chip.', 'Product_Color': 'Space Gray, Silver, Gold, Rose Gold', 'Product_Size': '13 inches', 'Product_Original_Price': '$999.00', 'Product_Discounted_Price': '$899.00', 'Product_Additional_Discount': '5%', 'Average_Rating': 4.8, 'Reviews': {'user1': "This laptop is amazing! It's so fast and powerful, and the battery life is incredible.", 'user2': "I love the design of this laptop. It's so sleek and stylish.", 'user3': "This laptop is a great value for the price. It's well-made and does everything I need it to do."}}"""

# Use eval() to safely evaluate the string into a dictionary
product_dict = eval(product_info)

# Print the resulting dictionary
print(product_dict)
```

```python
# alpaca_prompt = Copied from above
alpaca_prompt = f"""
### Instruction : Help user to solve is query
### Context : {product_dict}
### Input : Is the product worth buying
### Response : ''
"""
alpaca_prompt = alpaca_prompt.replace('{','[').replace('}',']')
alpaca_prompt = alpaca_prompt.replace('\n','')
inputs = tokenizer(alpaca_prompt, return_tensors="pt").input_ids
outputs = model.generate(inputs, max_length=350)
response = tokenizer.decode(outputs[0], skip_special_tokens=True)
response_start = "### Response :"

# Find the index where the response starts
start_index = response.find(response_start)

if start_index != -1:
    # Extract everything after the response start index
    extracted_response = response[start_index + len(response_start):].strip()
    print(extracted_response)
else:
    print("Response not found.")
```
