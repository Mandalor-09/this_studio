````python
import json
import pandas as pd

# Initialize an empty DataFrame
df = pd.DataFrame()

# List to store JSON data dictionaries (optional)
json_data_list = []

for _ in range(200):  # Change the range to generate more data if needed
    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[{"role": "user", "content": 'Create json data for the below function use random data' }],
        functions=generate_data
    )

    # Extract function call from the response
    function_call = response.choices[0].message.function_call

    # Parse the arguments JSON string
    arguments_dict = json.loads(function_call.arguments)

    # Add the data to the list (optional, for keeping track of JSON data)
    json_data_list.append(arguments_dict)

    # Convert arguments_dict to a DataFrame
    new_df = pd.DataFrame([arguments_dict])

    # Append the new DataFrame to the existing DataFrame
    df = pd.concat([df, new_df], ignore_index=True)

````

````python
import json
import time

conversation_list = []
last_processed_row = 0
````

````python
# Iterate over each chunk in the DataFrame 'df'
for chunk in df:
    df_chunk = chunk.iloc[:, 1:]  # Exclude the first column if needed

    sleep = time.sleep(60)

    # Iterate over each row in the current chunk, starting from last_processed_row
    for index, row in df_chunk.iloc[last_processed_row:].iterrows():
        try:
            print(index)
            data = row.to_json()  # Convert row to JSON format

            # Define the prompt for conversation script
            prompt = f"""**Product Information:** {data}

**Prompt:**

Create a conversation script for the given product information and return it in JSON format.

**Instructions:**

* Do not offer a discount greater than the product's `Product_Additional_Discount`.
* Aim to negotiate a lower price from the user's perspective.
"""
            #print(prompt)  # Print the prompt for debugging

            convo.send_message(prompt)

            data = convo.last.text

            # Preprocess received data (if needed)
            preprocessed_data = data.replace("```", "").replace('json', "").replace(">", ",").strip()

            # Load the preprocessed data into JSON format
            loaded_json = json.loads(preprocessed_data)

            # Append the loaded JSON to the conversation list
            conversation_list.append(loaded_json)

            # Update last_processed_row for successful processing
            last_processed_row = index + 1

        except Exception as e:
            print(f"Error processing row {index + last_processed_row}: {e}")
            # Handle the error (e.g., log it or retry)
            # You can decide on a retry strategy here (e.g., retry a limited number of times)
            break  # Exit the inner loop if an error occurs

print(f"Total conversations processed: {len(conversation_list)}")
````
